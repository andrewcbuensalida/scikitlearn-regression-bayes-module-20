````python
import numpy as np

def gradient_descent(
    gradient, start, learn_rate, n_iter=50, tolerance=1e-06
):
    vector = start
    for _ in range(n_iter):
        diff = -learn_rate * gradient(vector)
        #complete the if statement
        if abs(diff) <= tolerance:
            break
        vector += diff
    return vector
````

Now that you have a basic understanding of the implementation of a gradient descent function, it's time for you to test it.

Suppose that you want to find the local minimum of a $f(x) = x^2$ function. This function represents a parabola facing upwards.

Run the code cell below to visualize the $f(x)$ function.

````python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
y = x**2

plt.plot(x,y)
plt.plot(0,0, 'ro')
plt.title('Graph of a parabola')
````

In the figure above, the red dot represents the local minimum of the parabola, which in this case, is also called the vertex.

Now that you have defined the function, you can test the implementation of gradient descent by calling the `gradient_descent` *function*.

Note the following: You have set the `gradient` *argument* equal to $2x$ which represents the derivative of $f(x) = x^2$. The starting point, `start`, is set equal to 10 and the learning rate, `learn_rate`, is set equal to 0.2.

````python
gradient_descent(
    gradient=lambda v: 2 * v, start=0.4, learn_rate=0.2
)
````
